```python
import numpy as np
import nbinteract as nbi

def normal(mean, std):
    """Returns 1000 points drawn at random from N(mean, std)"""
    return np.random.normal(mean, std, 1_000)

nbi.hist(normal, mean=(0, 10), std=(0, 2.0))
```

```python
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

line, = ax.plot(x, np.sin(x))

def update(w=1.0):
    line.set_ydata(np.sin(w * x))
    fig.canvas.draw_idle()
    
interact(update);
```

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

def f(m, b):
    plt.figure(2)
    x = np.linspace(-15, 15, num=1000)
    plt.plot(x, m * x + b)
    plt.ylim(-5, 5)
    plt.show()

interactive_plot = interactive(f, m=(-2.0, 2.0), b=(-3, 3, 0.5))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```
